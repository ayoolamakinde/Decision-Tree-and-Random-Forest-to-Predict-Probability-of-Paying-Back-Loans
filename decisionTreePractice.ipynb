```python
#converting purpose to a category variable
purpose_cat = ['purpose']
loans_data = pd.get_dummies(loans,columns=purpose_cat,drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
X = loans_data.drop('not.fully.paid', axis=1)
y = loans_data['not.fully.paid']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

# Decision Tree 

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier()
tree.fit(X_train,y_train)
```

```python
tree_predictions = tree.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test,tree_predictions))
```

```python
print(classification_report(y_test,tree_predictions))
```

# Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(n_estimators=600)

forest.fit(X_train,y_train)
```

```python
forest_predictions = forest.predict(X_test)
```

```python
print(confusion_matrix(y_test,forest_predictions))
```

```python
print(classification_report(y_test,forest_predictions))
```
